```python
import warnings
warnings.simplefilter('ignore')

import matplotlib.pyplot as plt
import cv2 as cv
from face_verification.utils import *
from face_verification.api import *

%matplotlib inline
```

```python
img1 = read_image('img1.png')
img2 = read_image('img2.png')
img3 = read_image('img3.png')
img4 = read_image('img4.png')
```

```python
images = [img1,img2,img3,img4]
grays = []

for img in images:
    grays.append(BGR2Gray(img))
```

```python
# visualization
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(8, 8))
for i in range(4):
    ax[i // 2][i % 2].imshow(BGR2RGB(images[i]))
    ax[i // 2][i % 2].set_xticks([])
    ax[i // 2][i % 2].set_yticks([])
plt.tight_layout()
plt.show()
```

```python
# detect faces
haar_xml='./face_verification/xmls/haarcascade_frontalface_default.xml'
haar_faces = []
for image in images:
    try:
        faces = cascade_detector(image, xml=haar_xml, scale_factor=1.3, min_neighbors=5)
        haar_faces.append(faces)
    except:
        haar_faces.append([])
```

```python
# denote the faces with rectangles
rec_images = []
for i in range(len(images)):
    rec_images.append(denote_face(images[i], haar_faces[i]))

# visualization
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(8, 8))
for i in range(4):
    ax[i // 2][i % 2].imshow(BGR2RGB(rec_images[i]))
    ax[i // 2][i % 2].set_xticks([])
    ax[i // 2][i % 2].set_yticks([])
plt.tight_layout()
plt.show()
```

```python
# crop faces at scale_factor=1.0
crop_images = []
for i in range(len(images)):
    if len(haar_faces[i]) == 0:
        crop_images.append(images[i])
    else:
        cropped = crop_face(images[i], haar_faces[i], scale_factor=1.0, target_size=(128, 128))
        crop_images.append(cropped)

# visualization
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(6, 6))
for i in range(4):
    ax[i // 2][i % 2].imshow(BGR2RGB(crop_images[i]))
    ax[i // 2][i % 2].set_xticks([])
    ax[i // 2][i % 2].set_yticks([])
plt.tight_layout()
plt.show()
```

```python
# crop faces at scale_factor=1.3
crop_images = []
for i in range(len(images)):
    if len(haar_faces[i]) == 0:
        crop_images.append(images[i])
    else:
        cropped = crop_face(images[i], haar_faces[i], scale_factor=1.3, target_size=(128, 128))
        crop_images.append(cropped)

# visualization
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(6, 6))
for i in range(4):
    ax[i // 2][i % 2].imshow(BGR2RGB(crop_images[i]))
    ax[i // 2][i % 2].set_xticks([])
    ax[i // 2][i % 2].set_yticks([])
plt.tight_layout()
plt.show()
```

# Face Detection with LBP Cascade Detector

```python
# detect faces
lbp_xml='./face_verification/xmls/lbpcascade_frontalface_improved.xml'
lbp_faces = []
for image in images:
    try:
        faces = cascade_detector(image, xml=lbp_xml, scale_factor=1.1, min_neighbors=5)
        lbp_faces.append(faces)
    except:
        lbp_faces.append([])
```

```python
# denote the faces with rectangles
rec_images = []
for i in range(len(images)):
    rec_images.append(denote_face(images[i], lbp_faces[i]))

# visualization
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(8, 8))
for i in range(4):
    ax[i // 2][i % 2].imshow(BGR2RGB(rec_images[i]))
    ax[i // 2][i % 2].set_xticks([])
    ax[i // 2][i % 2].set_yticks([])
plt.tight_layout()
plt.show()
```
